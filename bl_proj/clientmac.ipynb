```python
import os
import sys
import threading
import time
import socket
import glob
import subprocess
import zipfile

debug = True

downloaded_files = []
uploaded_files = []
command_list = []
```

```python
#Looks for a broadcast message from the server on port 1000
#Uses this to determine the ip of the server
def lookforserver(port= 1000):
    if debug:
        print "Looking for command server"
        
    try:
        s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
        s.bind(('', port))

        data, addr = s.recvfrom(256)
    
        if debug:
            print "Server found at %s" % addr[0]
        return addr[0]
    except socket.error as msg:
        print msg, "Broadcast Server"
    except:
        print "Unknown error at broadcast server"
```

```python
#Data received on port 1001
#returns if files were successfully downloaded
#Automatically unzips file if file is a .zip file

def getdata(server_ip, port = 1001): 
    
    file_list = []
    
    
    if debug:
        print "Trying to download data"
    try:
        host = server_ip
        port = port
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

        s.connect((host,port))

        n = ord(s.recv(1)[0])
        
        if debug:
            print "Number of files to download: %d" % n
        file_list = []

        for i in range(n):
            filename, filesize = getfile(s)
            downloaded_files.append(filename)
            file_list.append(filename)
        
        s.close()
        return True
    except socket.error as msg:
        if debug:
            print msg, "Download server"
        return False
    except:
        if debug:
            print "Unknown error at download server"
        return False
    
    return True
    
    
#Get file with server connected on socket s
#First 128 bytes is simply a header
#returns filename and filesize
def getfile(s):
    if debug:
        print "Attempting download"
    data = s.recv(128)
    filename, filesize = decodeheader(data)
    
    if debug:
        print "Download file: %s\tFile size: %s" % (filename,filesize)
    
    f = open(filename, 'wb')
    
    percent = 0 
    dataread = 0
    while True: 
        if filesize - dataread >= 512:
            data = s.recv(512)
            dataread += len(data)
        else:
            data = s.recv(filesize-dataread)
            dataread = filesize
        
        if percent!=(dataread*100/filesize):
            percent = dataread*100/filesize
            print "%d%% download completed\r" % int(percent)
        if not data:
            break
        f.write(data)
        if dataread==filesize:
            break
    
    if debug:
        print "Download completed for file: %s" % filename
    
    f.close()
    fname,ext = os.path.splitext(filename)
    if ext == '.zip':
        if debug:
            print "Extracting file %s" % filename
        zipfile.ZipFile(filename).extractall()
            
    downloaded_files.append(filename)
    return filename, filesize

    
#In header ->
#First 120 bytes: File name
#Last 8 bytes: File size    
def decodeheader(data):
    filename = ''
    filesize = 0
    for i in range(0,120):
        filename = filename + data[i]
    
    for i in range(120,128):
        filesize = filesize*256 + ord(data[i])
    
    filename = filename.rstrip('\0')
    
    return filename, filesize
```

```python
#Uploads files in filelist with sever connected on socket conn
def uploadfile(conn, pathname):
    if debug:
        print "Starting a file upload procedure"

    filename = os.path.basename(pathname)
    size = os.path.getsize(pathname)

    if debug:
        print "Attempting to upload %s" % filename
    conn.send(encodeheader(filename, size))
    f = open(filename, 'rb')
    while True:
        data = f.read(512)
        if not data:
            break
        conn.send(data)
    f.close()
    if debug:
        print "%s uploaded" % filename
    uploaded_files.append(filename)
    

#Uploads data to server
#host = server ip
#default port = 1002
def uploaddata(host,filelist):
    port = 1002
    if debug:
        print "Attempting to start upload"
    try:

        print "File list is ", filelist
        
        for pathname in filelist:
            conn = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
            conn.connect((host,port))

            if debug:
                print "Connected to server"
            try:
                uploadfile(conn, pathname)
            except:
                if debug:
                    print "Unable to upload %s" % pathname
            finally:
                try:
                    conn.close()
                except:
                    if debug:
                        print "Unable to abort connection to server"
        return True
    except socket.error as msg:
        if debug:
            print msg, "Upload server"
        return False
    except:
        if debug:
            print "Unknown error in upload file function"
        return False
def encodeheader(filename, filesize):
    header = ''
    for i in range(len(filename)):
         header = header + filename[i]
    
    header = header + '\0'*(120-len(filename))
    
    size_str = ['\0']*8
    for i in range(7,-1,-1):
        if filesize == 0:
            break
        size_str[i] = chr(filesize%256)
        filesize/=256
    
    for i in range(0,8):
        header = header + size_str[i]
    return header[:128]
```

```python
#Executes a command
#The commands are differentiated on the basis of the first word
#download commands are executed by calling the getdata function
#upload commands are executed by calling the uploaddata function
#Other commands are run on the shell
def ExecuteCommand(cmd, serverip):
    #Cleaning cmd
    cmd.rstrip('\0')
    i = cmd.find('\x00')
    if not i==-1:
        cmd = cmd[:i]
    command = cmd.split()

    if command[0] == "download":
        getdata(serverip)
    elif command[0] == "upload":
        flist = command[1:]
        filelist = []
        for rawf in flist:
            for f in glob.glob(rawf):
                if not f in uploaded_files:
                    filelist.append(f)
        
        print "Uploading the following files: "
        print filelist
        
        return uploaddata(serverip, filelist)
        
    else:
        cmd = "/Applications/Blender/blender.app/Contents/MacOS/" + cmd
        if debug:
            print "Executing command: %s" % cmd
        return subprocess.call(cmd, shell=True)
    pass

#Runs a server which receives commands
#Will only execute the command if it is received from the serverip
#waitfunc called when server is waiting for a command
#execfunc called when server is starting to execute a command
class SlaveServer(threading.Thread):
    running = False
    def __init__(self, serverip):
        threading.Thread.__init__(self)
        if not SlaveServer.running:
            self.run_thr = True
            SlaveServer.serverip = serverip
            SlaveServer.running = True
        else:
            self.run_thr = False
    
    def run(self):
        if self.run_thr:
            serverip = SlaveServer.serverip
            if debug:
                print "Starting slave server for %s" % serverip
                sys.stdout.flush()
            try:
                s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
                s.bind(('', 1003))
                s.listen(1)

                if debug:
                    print "Waiting  for commands"
                    sys.stdout.flush()
                while True:
                    try:
                        self.waitfunc("waiting")
                    except:
                        if debug:
                            print "No external wait function present"
                    conn, addr = s.accept()
                    if debug:
                        print "Receiving command"
                    if addr[0] == serverip:
                        cmd = ''
                        while True:
                            data = conn.recv(512)
                            if not data:
                                conn.close()
                                break
                            cmd = cmd + data

                        try:
                            self.execfunc("executing")
                        except:
                            if debug:
                                print "No external exec function present"
                        print "Command server received:%s" % cmd
                        boolExecuted = ExecuteCommand(cmd, serverip)
                        command_list.append((cmd, boolExecuted))

                        if debug:
                            print "Command: ", cmd, "Status: ", boolExecuted

                    else:
                        if debug:
                            print "Fake master"
                        conn.close()
            except socket.error as msg:
                if debug:
                    print msg
            except Exception, e:
                if debug:
                    print e
                    print "Unknown error"
```

```python
#Constantly notifies server of the status of the client
#Server will declare timeout if the Heartbeat thread stops
class Heartbeat(threading.Thread):
    running = False
    status = ''
    def __init__(self, serverip, port=1004):
        threading.Thread.__init__(self)

        if not Heartbeat.running:
            self.run_thr = True
            Heartbeat.serverip = serverip
            Heartbeat.port = port
            Heartbeat.lock = threading.Lock()
            if debug:
                print "Stating heartbeats for %s" % serverip
            Heartbeat.running = True
        else:
            self.run_thr = False
            if debug:
                print "Cannot have more than one heart"
    
    def run(self):
        if self.run_thr:
            while True:
                self.beat()
                time.sleep(1)
    
    def beat(self):
        errorat = ''
        try:
            errorat = 'connecting'
            s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
            s.connect((Heartbeat.serverip, Heartbeat.port))
            errorat = 'sending'
            Heartbeat.lock.acquire()
            s.send(Heartbeat.status)
            Heartbeat.lock.release()
            errorat= 'closing'
            s.close()
            errorat = 'starting'
        except socket.error as msg:
            print msg, "Heartbeat server while %s" % errorat
        except:
            print "Unknown error at Heartbeat Server"
    
    def update(self, status):
        Heartbeat.lock.acquire()
        Heartbeat.status = status
        Heartbeat.lock.release()
        
        print "Updated status to: %s" % status
        self.beat()
        
```

```python
serverip = lookforserver()
SS = SlaveServer(serverip)
HB = Heartbeat(serverip)
SS.waitfunc = HB.update
SS.execfunc = HB.update
HB.start()
SS.start()
```
